抓三間公司的股價

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
```

```python
stock_nums = ["2357.TW", "2382.TW", "2417.TW"]
```

```python
# 股票號碼
stock_num:str = stock_nums[2]
# 資料開始日期
start_data = "2019-01-01"
# 資料結束日期
end_data = "2023-04-30"
```

```python
# driver物件
driver = webdriver.Chrome()

# 網址
url = "https://hk.finance.yahoo.com/"
# driver進入該網址
driver.get(url)

# 股票搜尋欄的HTML元素
stock_search_input = driver.find_element(by=By.XPATH, value="//*[@id=\"Col2-0-SymbolLookupV2-Proxy\"]/div/form/input")
# 將搜尋欄填入值(股票號碼)
stock_search_input.send_keys(stock_num)

time.sleep(1)

# 搜尋按鈕的HTML元素
stock_search_button = driver.find_element(by=By.XPATH, value="//*[@id=\"Col2-0-SymbolLookupV2-Proxy\"]/div/form/button[2]")
# 按下搜尋按鈕
stock_search_button.click()

# driver等待歷史紀錄的HTML元素載入網頁 timeout限制10秒
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id=\"quote-nav\"]/ul/li[4]/a"))
)

time.sleep(1)

# 歷史紀錄的HTML元素
history_data_button = driver.find_element(by=By.XPATH, value="//*[@id=\"quote-nav\"]/ul/li[4]/a")
# 按下歷史紀錄按鈕
history_data_button.click()

# driver等待日期選擇的HTML元素載入網頁 timeout限制10秒
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id=\"Col1-1-HistoricalDataTable-Proxy\"]/section/div[1]/div[1]/div[1]/div/div/div"))
)

time.sleep(1)

# 日期選擇的HTML元素
date_selecter = driver.find_element(by=By.XPATH, value="//*[@id=\"Col1-1-HistoricalDataTable-Proxy\"]/section/div[1]/div[1]/div[1]/div/div/div")
# 按下日期選擇按鈕
date_selecter.click()

time.sleep(1)

# 輸入開始日期的HTML元素
start_date_input = driver.find_element(by=By.NAME, value="startDate")

# 因為該元素操作起來有點小複雜 所以這裡不方便解釋 可以開啟網頁按F12看看
for data in start_data.split("-"):
    time.sleep(1)
    # 點一下鎖定
    start_date_input.click()
    # 先清除原本的值
    start_date_input.clear()
    # 輸入日期的年月日
    start_date_input.send_keys(data)
    # 模擬方向鍵右邊按一下 其實最後一個日輸入完後不用再按一下 不過我想讓code看起來乾淨一點 就不處理了
    start_date_input.send_keys(Keys.RIGHT)
    

# 結束日期 方法同上
end_date_input = driver.find_element(by=By.NAME, value="endDate")
for data in end_data.split("-"):
    time.sleep(1)
    end_date_input.click()
    end_date_input.clear()
    end_date_input.send_keys(data)
    end_date_input.send_keys(Keys.RIGHT)

time.sleep(1)

# 日期選擇完成按鈕的HTML元素
data_selecr_finish_button = driver.find_element(by=By.XPATH, value="//*[@id=\"dropdown-menu\"]/div/div[3]/button[1]")
# 按一下日期選擇完成按鈕
data_selecr_finish_button.click()

time.sleep(1)

# # 頻率選擇的HTML元素
# freq_selecter = driver.find_element(by=By.XPATH, value="//*[@id=\"Col1-1-HistoricalDataTable-Proxy\"]/section/div[1]/div[1]/div[3]/span/div")

# csv下載按鈕的HTML元素
data_download_button = driver.find_element(by=By.XPATH, value="//*[@id=\"Col1-1-HistoricalDataTable-Proxy\"]/section/div[1]/div[2]/span[2]/a")
# 按一下csv下載按鈕
data_download_button.click()

# 暫停3秒鐘 也可以放在其他地方慢慢觀察爬蟲過程
time.sleep(3)

# 關閉瀏覽器
driver.quit()
```
