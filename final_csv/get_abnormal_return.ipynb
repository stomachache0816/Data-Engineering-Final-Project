```python
import pandas as pd
import os
import statistics
```

```python
df = pd.DataFrame()
```

無風險利率

```python
file = pd.read_csv(filepath_or_buffer=os.getcwd() + "\\risk_free_interest_rate.csv")
df_risk_free = pd.DataFrame(file)
df["日期"] = df_risk_free["日期"]
df["無風險利率"] = df_risk_free["無風險利率"]
df
```

市場報酬率

```python
file = pd.read_csv(filepath_or_buffer=os.getcwd() + "\\index.csv")
df_index = pd.DataFrame(file)
df["市場報酬率"] = df_index["漲跌百分比"]
df
```

2357收盤價

```python
file = pd.read_csv(filepath_or_buffer=os.getcwd() + "\\price_2357.csv")
df_price_2357 = pd.DataFrame(file)
df["2357收盤價"] = df_price_2357["Close"]
df
```

2382收盤價

```python
file = pd.read_csv(filepath_or_buffer=os.getcwd() + "\\price_2382.csv")
df_price_2357 = pd.DataFrame(file)
df["2382收盤價"] = df_price_2357["Close"]
df
```

2417收盤價

```python
file = pd.read_csv(filepath_or_buffer=os.getcwd() + "\\price_2417.csv")
df_price_2417 = pd.DataFrame(file)
df["2417收盤價"] = df_price_2417["Close"]
df
```

2357實際報酬率

```python
L = [0]
prices = df["2357收盤價"]
for i in range(1, len(prices)):
    L.append((prices[i] - prices[i - 1]) / prices[i - 1])
df["2357實際報酬率"] = L
df
```

2382實際報酬率

```python
L = [0]
prices = df["2382收盤價"]
for i in range(1, len(prices)):
    L.append((prices[i] - prices[i - 1]) / prices[i - 1])
df["2382實際報酬率"] = L
df
```

2417實際報酬率

```python
L = [0]
prices = df["2417收盤價"]
for i in range(1, len(prices)):
    L.append((prices[i] - prices[i - 1]) / prices[i - 1])
df["2417實際報酬率"] = L
df
```

2357 Beta值

```python
x = df["2357實際報酬率"]
y = df["市場報酬率"]
beta_2357 = statistics.covariance(x, y) / statistics.variance(y)
beta_2357
```

2382 Beta值

```python
x = df["2382實際報酬率"]
y = df["市場報酬率"]
beta_2382 = statistics.covariance(x, y) / statistics.variance(y)
beta_2382
```

2417 Beta值

```python
x = df["2417實際報酬率"]
y = df["市場報酬率"]
beta_2417 = statistics.covariance(x, y) / statistics.variance(y)
beta_2417
```

2357超常報酬率

```python
L = [(df["2357實際報酬率"][i] - (df["無風險利率"][i] + (df["市場報酬率"][i] - df["無風險利率"][i]) * beta_2357)) for i in range(df.shape[0])]
df["2357超常報酬率"] = L
df
```

2382超常報酬率

```python
L = [(df["2382實際報酬率"][i] - (df["無風險利率"][i] + (df["市場報酬率"][i] - df["無風險利率"][i]) * beta_2382)) for i in range(df.shape[0])]
df["2382超常報酬率"] = L
df
```

2417超常報酬率

```python
L = [(df["2417實際報酬率"][i] - (df["無風險利率"][i] + (df["市場報酬率"][i] - df["無風險利率"][i]) * beta_2417)) for i in range(df.shape[0])]
df["2417超常報酬率"] = L
df
```

```python
df.to_csv(path_or_buf=os.getcwd() + "\\abnormal_return.csv", index=False)
```
