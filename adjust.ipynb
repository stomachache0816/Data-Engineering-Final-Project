```python
import pandas as pd
import os
```

為了做日期校正，我們要先把各個csv的日期陣列抓下來比對

其中一支股票價格(price)的日期

```python
file_price = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\price_datas\\2357.TW.csv", encoding="utf-8")
df_price = pd.DataFrame(file_price)
dates_price: list = list(df_price["Date"])
print(dates_price)
print(f"length = {len(dates_price)}")
```

其中一支股票外資佔比(invest)的日期

```python
file_invest = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\invest_datas\\invest_2357.csv", encoding="utf-8")
df_invest = pd.DataFrame(file_invest)
dates_invest = list(df_invest["日期"])
print(dates_invest)
print(f"length = {len(dates_invest)}")
```

大盤指數(index)的日期

```python
file_index = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\index_datas\\index.csv")
df_index = pd.DataFrame(file_index)
dates_index = list(df_index["日期"])
print(dates_index)
print(f"length = {len(dates_index)}")
```

無風險利率(risk-free interest rate)的日期

```python
file_risk_free = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\risk_free_interest_rate.csv")
df_risk_free = pd.DataFrame(file_risk_free)
dates_risk_free = list(df_risk_free["日期"])
print(dates_risk_free)
print(f"length = {len(dates_risk_free)}")
```

交集出一個大家都有的日期

```python
# 先將每個日期list轉成set 然後再用 & 交集，之後結果再轉回list。
dates_intersection_set: set = set(dates_price) & set(dates_index) & set(dates_invest) & set(dates_risk_free)
dates_intersection: list = list(dates_intersection_set)
dates_intersection.sort()
print(dates_intersection)
print(f"length = {len(dates_intersection)}")
```

準備要輸出csv的調整後的Dataframe

```python
file_price_2357 = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\price_datas\\2357.TW.csv")
df_price_2357 = pd.DataFrame(file_price_2357)
# 篩選出有交集出來的日期的列
df_price_2357_adj = df_price_2357.drop([i for i in range(df_price_2357.shape[0]) if df_price_2357["Date"][i] not in dates_intersection])
df_price_2357_adj["股票代號"] = ["2357"] * df_price_2357_adj.shape[0]
df_price_2357_adj
```

```python
file_price_2382 = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\price_datas\\2382.TW.csv")
df_price_2382 = pd.DataFrame(file_price_2382)
# 篩選出有交集出來的日期的列
df_price_2382_adj = df_price_2382.drop([i for i in range(df_price_2382.shape[0]) if df_price_2382["Date"][i] not in dates_intersection])
df_price_2382_adj["股票代號"] = ["2382"] * df_price_2357_adj.shape[0]
df_price_2382_adj
```

```python
file_price_2417 = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\price_datas\\2417.TW.csv")
df_price_2417 = pd.DataFrame(file_price_2417)
# 篩選出有交集出來的日期的列
df_price_2417_adj = df_price_2417.drop([i for i in range(df_price_2417.shape[0]) if df_price_2417["Date"][i] not in dates_intersection])
df_price_2417_adj["股票代號"] = ["2417"] * df_price_2417_adj.shape[0]
df_price_2417_adj
```

```python
file_invest_2357 = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\invest_datas\\invest_2357.csv")
df_invest_2357 = pd.DataFrame(file_invest_2357)
# 篩選出有交集出來的日期的列
df_invest_2357_adj = df_invest_2357.drop([i for i in range(df_invest_2357.shape[0]) if df_invest_2357["日期"][i] not in dates_intersection])
# 將百分比換成小數點。
df_invest_2357_adj["全體外資及陸資持股比率"] = [float(rate) * 0.01 for rate in df_invest_2357_adj["全體外資及陸資持股比率"]]
df_invest_2357_adj
```

```python
file_invest_2382 = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\invest_datas\\invest_2382.csv")
df_invest_2382 = pd.DataFrame(file_invest_2382)
# 篩選出有交集出來的日期的列。
df_invest_2382_adj = df_invest_2382.drop([i for i in range(df_invest_2382.shape[0]) if df_invest_2382["日期"][i] not in dates_intersection])
# 將百分比換成小數點。
df_invest_2382_adj["全體外資及陸資持股比率"] = [float(rate) * 0.01 for rate in df_invest_2382_adj["全體外資及陸資持股比率"]]
df_invest_2382_adj
```

```python
file_invest_2417 = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\invest_datas\\invest_2417.csv")
df_invest_2417 = pd.DataFrame(file_invest_2417)
# 篩選出有交集出來的日期的列。
df_invest_2417_adj = df_invest_2417.drop([i for i in range(df_invest_2417.shape[0]) if df_invest_2417["日期"][i] not in dates_intersection])
# 將百分比換成小數點。
df_invest_2417_adj["全體外資及陸資持股比率"] = [float(rate) * 0.01 for rate in df_invest_2417_adj["全體外資及陸資持股比率"]]
df_invest_2417_adj
```

```python
# 篩選出有交集出來的日期的列。
df_index_adj = df_index.drop([i for i in range(df_index.shape[0]) if df_index["日期"][i] not in dates_intersection])

cols = list(df_index_adj.columns)
cols[cols.index("漲跌百分比(%)")] = "漲跌百分比"
df_index_adj.columns = cols

# 將百分比換成小數點。 將 -- 或 --- 換成0
df_index_adj["漲跌百分比"] = [(0 if rate == "--" or rate == "---" else float(rate)) * 0.01 for rate in df_index_adj["漲跌百分比"]]

df_index_adj
```

```python
# 篩選出有交集出來的日期的列。
df_risk_free_adj = df_risk_free.drop([i for i in range(df_risk_free.shape[0]) if df_risk_free["日期"][i] not in dates_intersection])
# 將百分比換成小數點。
df_risk_free_adj["無風險利率"] = [rate * 0.01 for rate in df_risk_free_adj["無風險利率"]]
df_risk_free_adj
```
