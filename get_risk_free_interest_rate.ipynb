```python
import pandas as pd
import os
```

```python
file = pd.read_csv(filepath_or_buffer="臺灣30年期國債債券報酬率歷史數據.csv")
df = pd.DataFrame(file)
df
```

```python
cols: list = ["日期", "無風險利率"]
cols
```

```python
dates: list = list(df["日期"])

for i in range(len(dates)):
    year = dates[i].split("-")[0]
    month = dates[i].split("-")[1]
    day = dates[i].split("-")[2]
    
    month = f"{'0' if int(month) < 10 else ''}{month}"
    day = f"{'0' if int(day) < 10 else ''}{day}"
    
    dates[i] = f"{year}-{month}-{day}"
    
dates.reverse()
dates
```

```python
risks: list = list(df["收市"])
risks.reverse()
risks
```

```python
df_output = pd.DataFrame(data=[[dates[i], risks[i]] for i in range(df.shape[0])], columns=cols)
df_output
```

```python
df_output.to_csv(path_or_buf=f"{os.getcwd()}\\risk_free_interest_rate.csv", index=False)
```
