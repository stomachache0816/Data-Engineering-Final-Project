```python
import pandas as pd
import os
from sqlalchemy import create_engine
from sqlalchemy import NVARCHAR
from sqlalchemy import DATE
from sqlalchemy import FLOAT
```

股票資訊 Dataframe

```python
ids = ["2357", "2382", "2417"]
dfs = []

target_cols = ["證券代號", "證券名稱", "國際證券編碼"]
datas = []

for id in ids:
    file_invest = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\final_csv\\invest_{id}.csv")
    df_invest = pd.DataFrame(file_invest)
    datas.append(list(df_invest.iloc[0][target_cols]))

df_stock_info = pd.DataFrame(data=datas, columns=["股票代號", "股票名稱", "國際證券編碼"])
df_stock_info
```

股票日資料 Dataframe

```python
ids = ["2357", "2382", "2417"]
dfs = []

for id in ids:
    file_price = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\final_csv\\price_{id}.csv")
    df_price = pd.DataFrame(file_price)
    file_invest = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\final_csv\\invest_{id}.csv")
    df_invest = pd.DataFrame(file_invest)
    df = pd.DataFrame()
    # data=df_price[["股票代號", "Date", "Open", "High", "Low", "Close"]]
    df["股票代號日期"] = [f"{id}/{date}" for date in df_price["Date"]]
    df[["股票代號", "Date", "Open", "High", "Low", "Close"]] = df_price[["股票代號", "Date", "Open", "High", "Low", "Close"]]
    df["全體外資及陸資持股比率"] = df_invest["全體外資及陸資持股比率"]
    df["TR日期"] = [f"TR/{date}" for date in df_price["Date"]]
    df.columns = ["股票代號日期", "股票代號", "日期", "每日開盤價格", "每日最高價格", "每日最低價格", "每日收盤價格", "全體外資及陸資持股比率", "TR日期"]
    dfs.append(df)
    
df_stock_day_data = pd.concat(objs=dfs, axis=0, ignore_index=True)
df_stock_day_data
```

市場報酬率及無風險利率日資料 Dataframe

```python
file_index = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\final_csv\\index.csv")
df_index = pd.DataFrame(file_index)

file_risk_free = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\final_csv\\risk_free_interest_rate.csv")
df_risk_free = pd.DataFrame(file_risk_free)

df_index_and_risk_free =pd.DataFrame()
df_index_and_risk_free["TR日期"] = [f"TR/{date}" for date in df_index["日期"]]
df_index_and_risk_free["市場報酬率"] = df_index["漲跌百分比"]
df_index_and_risk_free["無風險利率"] = df_risk_free["無風險利率"]
df_index_and_risk_free["日期"] = df_index["日期"]
df_index_and_risk_free
```

股票月資料 DataFrame

```python
ids = ["2357", "2382", "2417"]
dfs = []

for id in ids:
    file_price = pd.read_csv(filepath_or_buffer=f"{os.getcwd()}\\price_month_datas\\{id}.TW.csv")
    df_price = pd.DataFrame(file_price)
    df = pd.DataFrame()
    df["股票代號日期"] = [f"{id}/{date}" for date in df_price["Date"]]
    df["股票代號"] = [id] * df_price.shape[0]
    df[["日期", "每月開盤價格", "每月最高價格", "每月最低價格", "每月收盤價格"]] = df_price[["Date", "Open", "High", "Low", "Close"]]
    dfs.append(df)
    
df_stock_month_data = pd.concat(objs=dfs, axis=0, ignore_index=True)
df_stock_month_data
```
