定義函式

```python
def get_df(target_index: list, date: str):
    url: str = f"https://www.twse.com.tw/rwd/zh/afterTrading/MI_INDEX?date={date}&type=ALL&response=json&_=1702368101002"
    response_obj = rq.request(method="GET", url=url)
    json_str: str = response_obj.text
    dic: dict = jsn.loads(s=json_str)
    if dic["stat"] == "很抱歉，沒有符合條件的資料!":
        return
    cols: list = dic["tables"][0]["fields"]
    datas: list = [dic["tables"][0]["data"][i] for i in range(len(dic["tables"][0]["data"])) if dic["tables"][0]["data"][i][0] in target_index]
    df = pd.DataFrame(data=datas, columns=cols)
    date:str = dic["date"]
    dates = [f"{date[:len('YYYY')]}-{date[len('YYYY'):len('YYYYMM')]}-{date[len('YYYYMM'):]}"] * df.shape[0]
    df["日期"] = dates
    df["漲跌(+/-)"] = [df["漲跌(+/-)"].iloc[i][-len("*</p>"):-len("</p>")] for i in range(len(df["漲跌(+/-)"]))]
    return df

def get_month_end_day(year: int, month: int):
    if month == 2:
        if year % 4 == 0 and not(year % 100 == 0) or year % 400 == 0:
            return 29
        else:
            return 28
    elif month in [1, 3, 5, 7, 8, 10, 12]:
        return 31
    else:
        return 30
```

```python
target_index = ["加權股價指數", "發行量加權股價指數"]

year = 2019
for month in range(1, 12 + 1):
    dfs = []
    for day in range(1, get_month_end_day(year=year, month=month) + 1):
        date = f"{year}{'0' if month < 10 else ''}{month}{'0' if day < 10 else ''}{day}"
        df_day = get_df(target_index=target_index, date=date)
        
        if df_day is not None:
            dfs.append(df_day)
            
        p = 4
        for i in range(p):
            print(f"剛爬完{date}的資料 休息中" + ("." * i + (" " * (p - i))), end="\r")
            time.sleep(0.25)
    
    df_month = pd.concat(dfs, axis=0, ignore_index=True)
    df_month.to_csv(path_or_buf=f"{os.getcwd()}\\index_month_datas\\index_{year}{'0' if month < 10 else ''}{month}.csv", encoding="UTF-8", index=False)
    
print(f"完成{year}年的資料爬取               ", end="\r")
```

```python
files = gb.glob(pathname=f"{os.getcwd()}\\index_month_datas\\*.csv", recursive=False)
dfs: list = []

for file in files:
    file_csv = pd.read_csv(filepath_or_buffer=file, encoding="UTF-8")
    df = pd.DataFrame(data=file_csv)
    dfs.append(df)
    
df_all = pd.concat(objs=dfs, axis=0)


path = os.getcwd() + "\\index_datas"
if not os.path.exists(path):
    os.makedirs(name=path)
file_name = f"index.csv"
file_path = path + "\\" + file_name
df_all.to_csv(path_or_buf=file_path, sep=",", encoding="UTF-8", index=False)
```
